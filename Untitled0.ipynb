```python
import numpy as np
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sahilislam007/college-student-placement-factors-dataset")

print("Path to dataset files:", path)
```

```python
dataset = pd.read_csv("/root/.cache/kagglehub/datasets/sahilislam007/college-student-placement-factors-dataset/versions/1/college_student_placement_dataset.csv")
```

```python
dataset['Internship_Experience'].value_counts()
dataset['Placement'].value_counts()
```

```python
dataset.replace(['Yes','No'],[1,0],inplace=True)
```

```python
dataset['Internship_Experience'].value_counts()
dataset['Placement'].value_counts()
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.drop(columns=['Placement'])
y = dataset['Placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prediction = model.predict(x_test)
acc = accuracy_score(y_test,prediction)
print(acc)
```
