```python
import os
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
```

```python
#0 = asian
#1 = african
```

```python


# Define paths
path_train_african = '/content/dataset/train/African/'
path_train_asian   = '/content/dataset/train/Asian/'
path_test_african  = '/content/dataset/test/African/'
path_test_asian    = '/content/dataset/test/Asian/'

# Load filenames
train_african = os.listdir(path_train_african)
train_asian   = os.listdir(path_train_asian)
test_african  = os.listdir(path_test_african)
test_asian    = os.listdir(path_test_asian)

# Data containers
data = []
labels = []

# Load and label images
for i in train_african:
    img = Image.open(os.path.join(path_train_african, i)).resize((224, 224)).convert('RGB')
    data.append(np.array(img))
    labels.append(1)

for i in train_asian:
    img = Image.open(os.path.join(path_train_asian, i)).resize((224, 224)).convert('RGB')
    data.append(np.array(img))
    labels.append(0)

for i in test_african:
    img = Image.open(os.path.join(path_test_african, i)).resize((224, 224)).convert('RGB')
    data.append(np.array(img))
    labels.append(1)

for i in test_asian:
    img = Image.open(os.path.join(path_test_asian, i)).resize((224, 224)).convert('RGB')
    data.append(np.array(img))
    labels.append(0)

print(f"Total images: {len(data)}, Total labels: {len(labels)}")
```

```python
data = np.array(data) / 255.0
labels = np.array(labels)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=2)
```

```python
model = keras.Sequential([
    keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Conv2D(64, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Flatten(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dropout(0.5),

    keras.layers.Dense(2, activation='softmax')  ])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
