```python
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
dataset = pd.read_csv('/content/parkinsons.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
x = dataset.drop(columns=['name','status'],axis=1)
y = dataset['status']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2 ,stratify =y,random_state=2)
```

```python
model = svm.SVC(kernel ='linear'  )
```

```python
model.fit(x_train,y_train)
```

```python
x_trian_prediction = model.predict(x_train)
acc = accuracy_score(y_train,x_trian_prediction)
print(acc)
```

```python
x_test_prediction = model.predict(x_test)
acc = accuracy_score(y_test,x_test_prediction)
print(acc)
```

```python
num = random.randint(1, 120)
new_data = x_test.iloc[num].values.reshape(1, -1)
predict = model.predict(new_data)
str = dataset['name'].iloc[num]
print(f'{str} having')
if predict == 0:
  print(f'{str} having No Disease')
else:
  print(f'{str} having Parkinsons Disease')
```
