```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/content/train-titanic.csv')
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.shape
```

```python
dataset = dataset.drop(columns='Cabin',axis =1)
```

```python
dataset['Age'] = dataset['Age'].fillna(dataset['Age'].mean())
```

```python
dataset['Fare'] = dataset['Fare'].fillna(dataset['Fare'].mean())
```

```python
dataset.isnull().sum()
```

```python
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0],inplace=True)
```

```python
dataset.shape
```

```python
dataset['Embarked'].value_counts()
```

```python
dataset  = dataset.replace({'Sex':{'male':1,'female':0},
                            'Embarked':{'S':0,'C':1,'Q':2}})
```

```python
dataset.columns
```

```python
x = dataset.drop(columns=['PassengerId','Survived','Name','Ticket'],axis=1)
y = dataset['Survived']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
acc = accuracy_score(y_train,x_train_prediction)
print(acc)
```

```python
x_test_prediction = model.predict(x_test)
acc = accuracy_score(y_test,x_test_prediction)
print(acc)
```
