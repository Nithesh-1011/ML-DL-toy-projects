```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/content/breast cancer prediction.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop(columns='Unnamed: 32' , axis =1)
```

```python
dataset.columns
```

```python
dataset.replace({ 'diagnosis':{'M':0,'B':1}},inplace=True)
```

```python
x = dataset.drop(columns=['id','diagnosis'],axis=1)
y = dataset['diagnosis']
```

```python
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,stratify=y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
acc = accuracy_score(x_train_prediction,y_train)
print(acc)
```

```python
x_test_prediction = model.predict(x_test)
acc = accuracy_score(x_test_prediction,y_test)
print(acc)
```

```python
new_data = x_test.iloc[10].values.reshape(1,-1)
prediction = model.predict(new_data)
if prediction[0] == 0:
  print('Malignant')
else:
  print('Benign')
```
